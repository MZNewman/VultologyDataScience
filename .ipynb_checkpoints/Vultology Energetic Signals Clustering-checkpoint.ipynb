- subsetting by Development

Granted, some are likely to yield far more useful insights than others due to having more natural clusters

There is also a choice of whether to cluster the samples themselves, or to cluster the signals.

### Clustering Types

We can also use at least two types of clustering algorithms to good effect:
- KMeans
- Hierarchical clustering

There is even more than that, and some are specialized in searching for outliers (like DBSCAN) but I have often found these a bit less interesting and less fruitful by comparison (which is no guarantee of course). The Hierarchical case is a little more complicated to interpret since the number of clusters is more unpredictable.

However, hierarchical clustering is better at discovering innate structure in a dataset than Kmeans clustering is, because hierarchical doesn't force a specific number of clusters (that doesn't necessarily make it better though).

KMeans is essentially answering the question: "If there WERE precisely 4 clusters in this data, what would they be, and are the 4 clusters we chose the most natural" (assuming we arbitrarily set the number to 4). And like I said, depending on how we cluster there are a lot of different answers and insights that could be gleaned.

There are also some methods to try different values of K and choose the best on for a given dataset, and they're surprisingly simple to use (one is called a Scree plot, which is often used with Principle Component Analysis).

### Scoring

There are a number of options for scoring clusters, but the simplest and most intuitive seem to be:
- RAND score (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.rand_score.html#:~:text=Rand%20index.,%5B1%5D%20%5B2%5D.), which essentially tells us how well our clusters stack up against the ground truth labels of CT theory
- Silhouette score (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html), which tells us how intrinsically well defined the clusters are (based on how dense the cluster is and how far apart it is from other clusters), entirely independent on what ground truth labels they are associated with in CT (which can potentially help us discover new patterns that CT doesn't already assume, by showing us well-defined clusters we weren't aware of). The only thing is that comparing the silhouette score between different types of clustering is not defined to work well, so it's a better idea to use it to compare clustering algorithms of the same type (e.g. Kmeans, Hierarchical, etc.)

### Miscellaneous Considerations

- If we cluster the signals, which seems like the motivation for this project, we have a huge number of dimensions (basically the sample size). This is a problem called the Curse of Dimensionality (very important in Machine Learning), which leads to overfitting the dataset and a lot of difficulty for clustering algorithms since more dimensions your space has (502 in our case), the more points appear equidistant to one another. This might require the use of dimensionality reduction techniques like PCA (Principal Component Analysis) for good performance on some clustering tasks.
- The problem with clustering samples instead is that it becomes potentially difficult to visualize so many samples unless we reduce them to examples of subclasses like Lead Energetic, Quadra, Development, and Type (but even Type is pretty difficult since that's 16 classes, it's a bit much for the eye, it's better to stay in a range of about 6/7 classes or less from a visual perspective), plus the samples don’t tell us as much abstractly as signals do
- We can also tweak the weights for signals strengths, since the weights of 2, 4 and 7 are technically arbitrary, and other values could potentially change the clusters somewhat. Signal weight can essentially be made into a hyperparameter, but this adds a lot of complexity
- I would also wonder if we will see some clusters associating closer to one another, like Je and Pi, which I would expect based on the Nemvus study on the Grant Stack.
- Maybe experiments could be designed where people would apply these signals without knowing about the assumptions of CT theory, but that seems complicated.

## K Means Clustering

### Clustering the 4 Energetics

We are clustering signals in this case, rather than samples

```python
k_values = range(2, 21)
k_values
```

```python
Energetics_samples = signals_df.iloc[:, np.r_[6:16, 21:31]]
Energetics_samples.columns, len(Energetics_samples.columns)
```

```python
Energetics_labels = [x[:2] for x in Energetics_samples.columns]
Energetics_labels
```

```python
Energetics = Energetics_samples.T
Energetics.index
```

```python
Energetics.head()
```

#### Testing

```python
kmeans = KMeans(n_clusters=4, random_state=42)
model = kmeans.fit(Energetics)
```

```python
clusters = model.predict(Energetics)
cluster_df = pd.DataFrame({'signal': Energetics.index, 'label': clusters})
cluster_df
```

#### Proper KMeans

```python
models = []
inertia = []

for k in k_values:
    model = KMeans(n_clusters=k, random_state=42).fit(Energetics)
    models.append(model)
    inertia.append(model.inertia_)
    
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia')
plt.title('Scree Plot for KMeans Clustering on 4 Energetics')
plt.xticks(k_values)
plt.show()
```

4 does seem to be the elbow, which is what we want

```python
elbow = models[2]
clusters = elbow.predict(Energetics)
cluster_df = pd.DataFrame({'signal': Energetics.index, 'label': clusters})
cluster_df
```

```python
rand_score(Energetics_labels, clusters)
```

```python
adjusted_rand_score(Energetics_labels, clusters)
```

```python
silhouette_score(Energetics, clusters)
```

```python
silhouette_df = pd.DataFrame({'signal': Energetics.index, 'score': silhouette_samples(Energetics, clusters)})
silhouette_df.sort_values('score')
```

This is very interesting in itself, but it would be even more interesting if we could see the silhouette score that each point would have in each cluster, so we can see what would be the 2nd greatest

```python
cluster_set = set(clusters)
scores = [[] for i in range(len(cluster_set))]

for i in range(len(Energetics)):
    for c in cluster_set:
        temp = clusters.copy()
        temp[i] = c
        temp_sil = silhouette_samples(Energetics, temp)
        scores[c].append(temp_sil[i])
        
mapping = {0: 'Pi', 1: 'Pe', 2: 'Ji', 3: 'Je'}
silhouette_df['cluster'] = cluster_df['label'].replace(mapping)

second_closest = []
differences = []

for i in range(len(scores[0])):
    sorted_scores = sorted(enumerate([scores[j][i] for j in range(len(scores))]), key=lambda x: x[1])
    second_closest.append(sorted_scores[-2][0])
    differences.append(sorted_scores[-1][1] - sorted_scores[-2][1])
    
silhouette_df['second_best'] = pd.Series(second_closest).replace(mapping)
silhouette_df['difference'] = pd.Series(differences)

silhouette_df
```

```python
silhouette_df = silhouette_df.merge(pd.DataFrame(scores).T, how='outer', left_index=True, right_index=True).rename(columns=mapping)
silhouette_df
```

```python
silhouette_df.sort_values('difference')
```

```python
perplexity = [2, 4, 6]

fig, axes = plt.subplots(len(perplexity), figsize=(12, 12*len(perplexity)))

for p, axis in zip(perplexity, axes):
    tsne = TSNE(n_components=2, perplexity=p, n_iter=1000, random_state=42, method='exact')
    embedded = pd.DataFrame(tsne.fit_transform(Energetics), index=silhouette_df.index)
    embedded['labels'] = silhouette_df['cluster'].astype('category')
    sns.scatterplot(ax=axis, x=embedded[0], y=embedded[1], data=embedded, hue='labels', palette=sns.color_palette(n_colors=len(embedded['labels'].unique())))

plt.show()
```

```python
perplexity = [2]

fig, axes = plt.subplots(len(perplexity), figsize=(12, 12*len(perplexity)))

tsne = TSNE(n_components=2, perplexity=perplexity[0], n_iter=1000, random_state=42, method='exact')
embedded = pd.DataFrame(tsne.fit_transform(Energetics), index=silhouette_df.index)
embedded['labels'] = silhouette_df['cluster'].astype('category')
sns.scatterplot(ax=axes, x=embedded[0], y=embedded[1], data=embedded, hue='labels', palette=sns.color_palette(n_colors=len(embedded['labels'].unique())))
for i, txt in enumerate(Energetics.index):
    plt.text(embedded.iloc[i, 0], embedded.iloc[i, 1], txt, fontsize=8)

plt.show()
```

These calculations of z-scores are incorrect, I'm going to recheck it manually

#### Recheck: Testing Shoulder Shrugs against Lead Functions

```python
PR3_LF_mean = signals_df['PR-3 Shoulder Shrugs:'].mean()
PR3_LF_std = signals_df['PR-3 Shoulder Shrugs:'].std()
PR3_LF_submeans = signals_df.groupby('Lead Function')['PR-3 Shoulder Shrugs:'].mean().sort_values(ascending=False)
PR3_LF_subcounts = signals_df.groupby('Lead Function')['PR-3 Shoulder Shrugs:'].count().sort_values(ascending=False)
PR3_LF_mean, PR3_LF_std, PR3_LF_submeans, PR3_LF_subcounts
```

```python
functions = list(PR3_LF_submeans.index)
z_scores = []

for f in functions:
    zscore = (PR3_LF_submeans[f] - PR3_LF_mean)*np.sqrt(PR3_LF_subcounts[f]) / PR3_LF_std
    z_scores.append(zscore)
    
pd.DataFrame(zip(functions, z_scores))
```

#### Recheck: Testing Shoulder Shrugs against Quadrant Signals

```python
PR3_PQS_mean = signals_df['PR-3 Shoulder Shrugs:'].mean()
PR3_PQS_std = signals_df['PR-3 Shoulder Shrugs:'].std()
PR3_PQS_submeans = signals_df.groupby('P Quadra Signal')['PR-3 Shoulder Shrugs:'].mean().sort_values(ascending=False)
PR3_PQS_subcounts = signals_df.groupby('P Quadra Signal')['PR-3 Shoulder Shrugs:'].count().sort_values(ascending=False)
PR3_PQS_mean, PR3_PQS_std, PR3_PQS_submeans, PR3_PQS_subcounts
```

```python
pq_signals = list(PR3_PQS_submeans.index)
z_scores = []

for s in pq_signals:
    zscore = (PR3_PQS_submeans[s] - PR3_PQS_mean)*np.sqrt(PR3_PQS_subcounts[s]) / PR3_PQS_std
    z_scores.append(zscore)
    
pd.DataFrame(zip(pq_signals, z_scores))
```
